```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
School_count = len(school_data_complete['School ID'].unique())
Student_count = len(school_data_complete['Student ID'].unique())
Total_budget = school_data['budget'].sum()
AvgMathScore = student_data['math_score'].mean()
AvgReadingScore = student_data['reading_score'].mean()
Passing_math = ((len(student_data.loc[student_data["math_score"] >= 70,:])/Student_count)*100)
Passing_reading = ((len(student_data.loc[student_data["reading_score"] >= 70,:])/Student_count)*100)
Overall_Passing = ((len(student_data.loc[(student_data["math_score"] >= 70)&(student_data["reading_score"] >= 70),:])/Student_count)*100)
Final_df = pd.DataFrame({"Total Schools": [School_count], "Total Students": [Student_count], "Total Budget": [Total_budget], "Average Math Score": [AvgMathScore], "Average Reading Score":[AvgReadingScore], "% Passing Math":[Passing_math], "% Passing Reading":[Passing_reading], "% Overall Passing":[Overall_Passing]})
Final_df['Total Budget'] = Final_df['Total Budget'].astype(float)
cols = ["Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"]
Final_df[cols] = Final_df[cols].applymap(lambda x: '{0:.4f}'.format(x))
Final_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
SchoolType = school_data_complete.loc[:,["school_name",'Student ID','reading_score','math_score','budget','type']]
SchoolType_group = SchoolType.groupby('school_name')
Student_count = SchoolType_group['Student ID'].count().to_frame()
Student_count['School Type'] = SchoolType_group['type'].unique()
Total_School_budget = SchoolType_group['budget'].sum().to_frame()
Student_count['Total School Budget'] = Total_School_budget['budget']/Student_count['Student ID']
Student_count['Per Student Budget'] = Student_count['Total School Budget']/Student_count['Student ID']
Student_count['Average Math Score'] = SchoolType_group['math_score'].mean().astype(float).map("{:,.4f}".format)
Student_count['Average Reading Score'] = SchoolType_group['reading_score'].mean().astype(float).map("{:,.4f}".format)
Passing_math = SchoolType.loc[SchoolType["math_score"] >= 70,:]
Passing_math_school = Passing_math.groupby('school_name')
Passing_math_count = Passing_math_school['Student ID'].count().to_frame()
Passing_reading = SchoolType.loc[SchoolType["reading_score"] >= 70,:]
Passing_reading_school = Passing_reading.groupby('school_name')
Passing_reading_count = Passing_reading_school ['Student ID'].count().to_frame()
Passing_overall = SchoolType.loc[(SchoolType["reading_score"] >= 70) & (SchoolType["math_score"] >= 70),:]
Passing_overall_school = Passing_overall.groupby('school_name')
Passing_overall_count = Passing_overall_school ['Student ID'].count().to_frame()
Student_count['% Passing Math'] = ((Passing_math_count['Student ID'] / Student_count['Student ID'])*100).astype(float).map("{:,.6f}".format)
Student_count['% Passing Reading'] = ((Passing_reading_count['Student ID'] / Student_count['Student ID'])*100).astype(float).map("{:,.6f}".format)
Student_count['% Passing Overall'] = ((Passing_overall_count['Student ID'] / Student_count['Student ID'])*100).astype(float).map("{:,.6f}".format)
Cols = Student_count.columns.tolist()
Cols = ['School Type',
 'Student ID',
 'Total School Budget',
 'Per Student Budget',
 'Average Math Score',
 'Average Reading Score',
 '% Passing Math',
 '% Passing Reading',
 '% Passing Overall']
Student_count = Student_count[Cols]
Final_df_School = Student_count.rename(columns={"Student ID":"Total Students"})
del Final_df_School.index.name
Final_df_School
```

## Reading Score by Grade 

```python
MathGrade = school_data_complete.loc[:,["school_name",'Student ID','math_score','grade']]

MathGrade_9th = MathGrade.loc[MathGrade["grade"] == "9th",:]
MathGrade_9th_Group = MathGrade_9th.groupby('school_name')
MathGrade_9th_Ser = MathGrade_9th_Group.mean()

MathGrade_10th = MathGrade.loc[MathGrade["grade"] == "10th",:]
MathGrade_10th_Group = MathGrade_10th.groupby('school_name')
MathGrade_10th_Ser = MathGrade_10th_Group.mean()

MathGrade_11th = MathGrade.loc[MathGrade["grade"] == "11th",:]
MathGrade_11th_Group = MathGrade_11th.groupby('school_name')
MathGrade_11th_Ser = MathGrade_11th_Group.mean()

MathGrade_12th = MathGrade.loc[MathGrade["grade"] == "12th",:]
MathGrade_12th_Group = MathGrade_12th.groupby('school_name')
MathGrade_12th_Ser = MathGrade_12th_Group.mean()

Final_MathGrade_df = pd.DataFrame(MathGrade_9th_Ser['math_score'])
Final_MathGrade_df['10th'] = MathGrade_10th_Ser['math_score']
Final_MathGrade_df['11th'] = MathGrade_11th_Ser['math_score']
Final_MathGrade_df['12th'] = MathGrade_12th_Ser['math_score']
Final_MathGrade_df = Final_MathGrade_df.rename(columns={'math_score':'9th'})
del Final_MathGrade_df.index.name
pd.options.display.float_format = '{:,.6f}'.format
Final_MathGrade_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
SchoolSpending = school_data_complete.loc[:,['school_name','Student ID','reading_score','math_score','budget']]
SchoolGrouped = SchoolSpending.groupby('school_name')
Student_count_2 = SchoolGrouped['Student ID'].count().to_frame()
Total_School_budget = SchoolGrouped['budget'].sum().to_frame()

Student_count_2['Total School Budget'] = Total_School_budget['budget']/Student_count_2['Student ID']
Student_count_2['Per Student Budget'] = Student_count_2['Total School Budget'].astype('int32')/Student_count_2['Student ID'].astype('int32')
Student_budget = Student_count_2.loc[:,['Per Student Budget']]
Results = pd.merge(SchoolSpending, Student_budget, on='school_name')
Result_df_Final = Results.loc[:,['Per Student Budget','reading_score','math_score']]
Result_df_Final = Result_df_Final.rename(columns={'Per Student Budget':'Spending Ranges (Per Student)'})
SS_bins = [0,583,630,645,676]
SS_labels = ['<584','$585-629','$630-644','$645-675']
Result_df_Final_Cut = pd.cut(Result_df_Final['Spending Ranges (Per Student)'], bins=SS_bins, labels=SS_labels)
Result_df_Final_df = pd.DataFrame(SchoolSpending.groupby(Result_df_Final_Cut)['math_score'].mean().to_frame())
Result_df_Final_df['Average Reading Score'] = SchoolSpending.groupby(Result_df_Final_Cut)['reading_score'].mean().to_frame()

Passing_math = Result_df_Final.loc[Result_df_Final["math_score"] >= 70,:]
Passing_reading = Result_df_Final.loc[Result_df_Final["reading_score"] >= 70,:]
Passing_overall = Result_df_Final.loc[(Result_df_Final["reading_score"] >= 70) & (Result_df_Final["math_score"] >= 70),:]

Passing_math_Cut = pd.cut(Passing_math['Spending Ranges (Per Student)'], bins=SS_bins, labels=SS_labels)
Passing_reading_Cut = pd.cut(Passing_reading['Spending Ranges (Per Student)'], bins=SS_bins, labels=SS_labels)
Passing_overall_Cut = pd.cut(Passing_overall['Spending Ranges (Per Student)'], bins=SS_bins, labels=SS_labels)


Total_count = Result_df_Final.groupby(Result_df_Final_Cut)['math_score'].count().to_frame()
PassingMath_Count = Passing_math.groupby(Passing_math_Cut)['math_score'].count().to_frame()
Passing_reading_Count = Passing_reading.groupby(Passing_reading_Cut)['math_score'].count().to_frame()
Passing_overall_Count = Passing_overall.groupby(Passing_overall_Cut)['math_score'].count().to_frame()

Result_df_Final_df['% Passing Math'] = ((PassingMath_Count['math_score']/Total_count['math_score'])*100)
Result_df_Final_df['% Passing Reading'] = ((Passing_reading_Count['math_score']/Total_count['math_score'])*100)
Result_df_Final_df['% Overall Passing'] = ((Passing_overall_Count['math_score']/Total_count['math_score'])*100)
Result_df_Final_df = Result_df_Final_df.rename(columns={"math_score":"Average Math Score"})
Result_df_Final_df
```
